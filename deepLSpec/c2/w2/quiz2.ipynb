```python
import math
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
a0 = 0.1
t = range(1, 50)
t = np.arange(0.1, 0.9, 0.05)
```

```python
plt.plot(t, [math.e**t_i * a0 for t_i in t])

plt.title('a = e**t * a0\n')
plt.xlabel('epoch')
plt.ylabel('learning rate')
plt.show()
print('opposite of what we want. terrible!')
```

```python
plt.plot(t, [1/math.sqrt(t_i) * a0 for t_i in t])

plt.title('a = 1/sqrt(t) * a0\n')
plt.xlabel('epoch')
plt.ylabel('learning rate')
plt.show()
```

```python
plt.plot(t, [0.95**t_i * a0 for t_i in t])

plt.title('a = 0.95**t_i * a0\n')
plt.xlabel('epoch')
plt.ylabel('learning rate')
plt.show()
```

```python
plt.plot(t, [1/(1+2*t_i) * a0 for t_i in t])

plt.title('a = 1/(1+2*t_i) * a0\n')
plt.xlabel('epoch')
plt.ylabel('learning rate')
plt.show()
```
